## Scaling

```python
img = cv.imread("D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/logo2.jpg")
print(img.shape)

resized_img = cv.resize(img, (150, 150), interpolation=cv.INTER_LINEAR)
print(resized_img.shape)
plt.imshow(cv.cvtColor(resized_img, cv.COLOR_BGR2RGB))
plt.show()

resized_img = cv.resize(img, None, fx=0.5, fy=0.5, interpolation=cv.INTER_LINEAR)
print(resized_img.shape)
plt.imshow(cv.cvtColor(resized_img, cv.COLOR_BGR2RGB))
plt.show()
```

##  Translation

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/car1.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

rows,cols = img.shape
 
M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap="gray")
plt.show()
```

## Rotation

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/car1.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

rows,cols = img.shape
 

# center, degrees, scale factor
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),73,1)
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap="gray")
plt.show()
```

## Affine Transformation

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/car1.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

rows,cols,ch = img.shape
 
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
 
M = cv.getAffineTransform(pts1,pts2)
 
dst = cv.warpAffine(img,M,(cols,rows))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Perspective Transformation

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/sudoku.jpg')
rows,cols,ch = img.shape
 
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
 
M = cv.getPerspectiveTransform(pts1,pts2)
 
dst = cv.warpPerspective(img,M,(300,300))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Задание 2

## 2D Convolution ( Image Filtering ) 

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/logo.jpg')

kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Image Blurring (Image Smoothing) 

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/sudoku.jpg')

blur = cv.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Gaussian Blurring

```python
img = cv.imread('D:/PyProjects/SLAM_Algorithms/SLAM-Algorithms/Lab_3/data/sudoku.jpg')

blur = cv.GaussianBlur(img, (7, 7), 30)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
