## Задача 1

Построить график рыночного равновесия на холодильники в
магазине за день. Определите равновесную цену (Pе) и равновесный объем
продаж (Qе). Определить наличие дефицита и избытка товара при ценах 100
и 400 руб.

Функции:
1. Функция спроса: Qd= 900 – Р.
2. Функция предложения: Qs= 100 + 3Р.

```python
p = [x for x in range(1, 900)]
qd = [900 - x for x in p]
qs = [100 + 3 * x for x in p]
plt.plot(qd, p, label="Спрос")
plt.plot(qs, p, label="Предложение", color='r')
plt.scatter(700, 200, label="Рыночное равновесие", color='r')
plt.legend()
plt.xlabel("Q, количество товара")
plt.ylabel("Р, Цена")
plt.show()
```

900 - P = 100 + 3P\
4P = 800\
P = 200 - Равновесная цена\
Q = 900-200=100+3*200=700 - Равновесный объем


При P = 100 - Дефицит = 800 - 400 = 400\
При P = 400 - Избыток = 1300 - 500 = 800

## Задача 3

При повышении цены на молоко с 30 до 35 руб. за 1 литр в
магазине объём спроса на него сократился с 100 до 98 л. Определите тип
эластичности спроса на молоко, изменение общей выручки продавца.

```python
P1 = 30
P2 = 35
Q1 = 100
Q2 = 98
E = (Q2 - Q1)/(P2 - P1) * (P1 + P2)/(Q1 + Q2)
E = abs(E)
print(f"{E:.2f}%")
```

|E| < 1% -> товар неэластичный

```python
delta = P1 * Q1 - P2 * Q2
abs(delta)
```

## Задача 8

Определите производительность труда (В) по предприятию в
целом и отдельно по каждому виду продукции предприятия, если известно, что
предприятие выпускает два вида изделий (А и Б); объем валовой продукции (ВП)
изделия А составляет 13 000 тыс. р., Б – 28 000 тыс. р., на предприятии работает
(Чппп) 3 670 человек, из них в производстве продукции А занято 26 %

```python
a = 13000
b = 28000
h = 3670
h1 = int(h * 0.26)
h2 = h - h1
v = (a + b) / h
v1 = a / h1
v2 = b / h2
print(f"Производительность труда предприятия: {v:0.2f}")
print(f"Производительность труда предприятия при производстве товара А: {v1:0.2f}")
print(f"Производительность труда предприятия при производстве товара Б: {v2:0.2f}")
```

## Задача 9

реднесписочная численность работников предприятия за год
составила 900 человек. За период на предприятие поступило 49 человек, уволилось
по собственному желанию 52 человека, уволены за нарушение трудовой
дисциплины 43 человека, ушли на пенсию 21 человек. Определите коэффициент
текучести кадров (kт.к), коэффициент общего оборота кадров (kо.к), коэффициент
оборота по приему кадров (kп.к), коэффициент оборота по выбытию кадров (kв.к).

```python
h = 900
new = 49
left = 52
fired = 43
old = 21
ktk = (left+fired) / h * 100
kok = (new - (left+fired+old)) / h
kpk = new/h * 100
kvk = (left+fired+old) / h * 100
print(f"Коэффициент текучести кадров: {ktk:0.2f}%")
print(f"Коэффициент общего оборота кадров: {kok:0.2f}%")
print(f"Коэффициент оборота по приему кадров: {kpk:0.2f}%")
print(f"Коэффициент оборота по выбытию кадров: {kvk:0.2f}%")
```
